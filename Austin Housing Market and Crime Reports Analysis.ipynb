# About

Austin, the capital city of the U.S. State of Texas, was incorporated on December 27, 1839 and is now considered the 11th-most populous city. It has also been one the fastest growing and with that means a fast growing housing market. The Austin City government's goal in the last decade has been to make it a mission to "be the most livable city in the U.S."  This is accomplished with the 'Imagine Austin' iniative to implement goals to maintain affordable housing.

In 2014, the City published a comprehensive study on its Housing Market for Home Owners and Renters with kinds of useful and interesting information on demographics, total households vs. renters, and income growth. Specifically I wanted to focus in the top buying factors between Inner City and Commuting Austinites the decisions they weigh in purchasing.

While cost is a primary concern for both buyer types, Commuters in particular listed 'low crime/safe' as their second biggest concern while Inner City buyers listed 'neighborhood'. This Project focuses on joining free data sets from Houses Sold in the Austin area and compares them with Crime Reports by Zipcode to ultimately create a 'safety' ranker and find other interesting insights.

Using the following tools: SQL, Python, and Tableau I aim to find the safest Neighborhoods and Regions in Austin.

# Additional Analysis

While completing this project and looking at the density of crime 'hotspots' per zipcode, I noticed that that typically listings are priced much lower in these neighborhoods. Going in-depth would be beyond the scope of this project, since could be any number of factors.

If I had more data it might be possible to assess other information:

1) Why are the factors causing more crimes in specific neighborhoods?

One particular hypothesis from this data could be the following: correlation between the price range in a region compared with the number of crime occurrences. To see if there might be, check the heat map below.

```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np
from datetime import datetime
from dateutil.parser import parse
import re
import timeit

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data
df = pd.read_csv('TableauImportAustin10_AustinJoin(InnerCity).csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# List data types for columns

df.dtypes
```

```python
# Need to test correlation points. Are there are any correlations with Budget/Company and Gross?
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['CrimeOccurrencesAtZip'], y=df['latestPrice'])

plt.title('Sold Price vs. Crime Occurrences')

plt.xlabel('Sold Price')
plt.ylabel('Crime Occurrences')

plt.show()
```

```python
df.corr(method = 'pearson') #methods -> pearson(default), kendall, spearman
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
# numerize the columns

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#try correlation again

correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
# let's drop some columns

df_dropped = df.drop(['latitude', 'longitude','latest_saledate','latest_salemonth','full_address','zipcode.1','description'], axis=1)
```

```python
#try correlation again

correlation_matrix2 = df_dropped.corr(method = 'pearson')

sns.heatmap(correlation_matrix2, annot=True)

plt.show()
```

```python
#unstack to find highest pairs

correlation_matrix2.corr()

corr_pairs = correlation_matrix2.unstack()

corr_pairs
```

```python
#sort them

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#show highest
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
